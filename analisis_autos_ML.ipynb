# Tratamiento de datos

**En esta seccion se espera realizar:**
1. Cambiar el nombre de las columnas a español.
2. Cambiar el tipo de dato de las columnas del dataFrame.
3. Realizar correcciones en las cadenas de texto de los registros.
4. Crear columnas que sean pertinentes para el análisis de datos.
5. Eliminar columnas que no sean pertinentes para el análisis.
6. Tratar datos nulos.

```python
# Cambiar nombre de columnas a español
```

```python
auto.rename(columns = {
    'money':'Precio',
    'brand': 'Marca',
    'model': 'Modelo',
    'year': 'Anio',
    'color': 'Color',
    'fuel_type': 'Tipo_Combustible',
    'door': 'Puertas',
    'gear': 'Transmision',
    'motor': 'Motor',
    'body_type': 'Carroceria',
    'kilometres': 'Kilometros',
    'currency': 'Moneda'
}, inplace = True)
```

```python
auto
```

```python
# Cambiar tipo de datos
```

```python
auto = auto.astype({
    'Marca':'string',
    'Modelo':'string',
    'Color':'string',
    'Tipo_Combustible': 'string',
    'Puertas':'int',
    'Transmision': 'string',
     'Motor':'string',
    'Carroceria': 'string',
    'Moneda': 'string'
})
```

```python
# Cambiar cadena de caracteres de la columna Moneda.
```

```python
auto['Moneda']= auto['Moneda'].replace('pesos','ARS')
auto['Moneda']= auto['Moneda'].replace('dólares','USD')
```

```python
# Crear columna del precio de auto en dólares.
```

```python
auto['Precio_Dolar'] = None
```

```python
# Asignar valores a la columna Precio_Dolar. Los precios expresados en dólares que el vendedor publicó se mantendrán.
```

```python
# Valor dolar paralelo (10/01/2023): $357,00
valor_dolar={
    'ARS':357,
    'USD':1,
}

for moneda in valor_dolar.keys():
    auto.loc[ auto['Moneda'] == moneda , 'Precio_Dolar' ] = auto.loc[auto['Moneda']== moneda , 'Precio'] // valor_dolar[moneda]
auto['Precio_Dolar'] = auto['Precio_Dolar'].astype('int')
```

```python
#Se eliminará la columna Precio y Moneda.
```

```python
auto.drop('Precio', axis = 1, inplace=True )
auto.drop('Moneda', axis = 1, inplace=True )
```

```python
auto
```

```python
# Tratamiento de datos nulos.
```

```python
# Conteo de las columnas con valores nulos.
auto.isna().sum()
```

```python
# Se asignara valor: Desconocido, a cada registro que posea valor nulo.
```

```python
auto.fillna('Desconocido', inplace= True)
```

```python
auto.isna().sum()
```
